```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.normal(loc=50.0, scale=20.0, size=100)
x
```

```python
y=np.random.normal(loc=50.0, scale=20.0, size=100)
y
```

```python
plt.scatter(x, y)
plt.title("(1).Plot the values of x and y")
plt.xlabel("Office size")
plt.ylabel("Office price")
plt.show()
```

Mean squared error

```python
#Initialize m and c
m=10
c=2
#Find y_hat values
y_hat=m*x + c
#Get the difference between the y and y hat values
difference_array = np.subtract(y, y_hat)
#Square the difference
squared_array=difference_array*difference_array
#Sum the squared values 
sum=0
length=len(squared_array)
for i in range(length):
  sum=sum+squared_array[i]

#Get the mean
mean_squared_error=sum/length
mean_squared_error
```
