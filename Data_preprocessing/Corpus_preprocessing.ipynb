```
import nltk
from nltk.tokenize import sent_tokenize #To uniquely get sentences from running text (nlp library)
from sklearn.feature_extraction.text import TfidfVectorizer #machine learning libraries
import pandas as pd
```

```
nltk.download('punkt')
```

```
raw_text=open('corpus.txt')
```

```
corpus=sent_tokenize(raw_text.read())
```

```
#corpus
```

```
vectorizer=TfidfVectorizer()
```

```
X=vectorizer.fit_transform(corpus)
```

```
print(X)
```

```
vocab=vectorizer.get_feature_names() #Prints the words/features/vocabulary in the document
```

```
X.shape #Total number of columns and vocabularies
```

```
#Grab the first sentence(SENTENCES AS DOCUMENTS) and store in a variable
one_vector=X[0]
print(one_vector)
```

```
#Pick the vector,transpose it then make it denser
df=pd.DataFrame(one_vector.T.todense(),index=vocab,columns=['tfidf'])
#Sort with first 10 values acsending(head(10) or last 10 values(tail(10)))
df.sort_values(by=['tfidf'],ascending=False).head(10)
```
