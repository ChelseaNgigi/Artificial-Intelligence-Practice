Doing library imports

```python
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Grab the dataset

```python
our_fashion_dataset=keras.datasets.fashion_mnist
(train_images,train_labels),(test_images,test_labels)=our_fashion_dataset.load_data()
```

```python
class_names=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
```

Viewing the train dataset

```python
plt.imshow(train_images[5])
print(class_names[train_labels[5]])
```

Preparing the model

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
print(model.summary())
```

Setting up the model for training

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['acc'])
```

Train the model

```python
model.fit(train_images,train_labels,epochs=20)
```

Testing the model

```python
_,accuracy=model.evaluate(test_images,test_labels)
```

```python
accuracy
```

Use the model

```python
predictions=model.predict(test_images)
```
