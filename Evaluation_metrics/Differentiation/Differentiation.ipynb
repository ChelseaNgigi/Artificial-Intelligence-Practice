```python
import tensorflow as tf
from sympy import *
import matplotlib.pyplot as plt
```

Automatic differentiation

```python
#x=tf.Variable(3.)
#x=tf.Variable([1.,2.,5.])
x=tf.linspace(-8.,10.,1000)
```

```python
with tf.GradientTape() as tape:
  #y=3*x**5
  tape.watch(x)
  #y=tf.nn.tanh(x)
  #y=tf.nn.sigmoid(x)
  y=tf.nn.relu(x)
```

```python
dy_dx=tape.gradient(y,x)
```

```python
plt.plot(x,y,label='tanh')
plt.plot(x,dy_dx,label='dy_dx')
plt.legend()
```

```python
dy_dx.numpy()
```
